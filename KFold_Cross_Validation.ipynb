```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
```

## https://gist.github.com/ktisha/c21e73a1bd1700294ef790c56c8aec1f

```python
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
df=pd.read_csv("/content/pima-indians-diabetes.csv",names=names,skiprows=1)
```

```python
df
```

```python
X=df.drop('class',axis=1)
y=df['class']
```

```python
X
```

```python
y
```

```python
k=50 ## Ideally can be 10
seed=7
kfold=KFold(n_splits=k,shuffle=True,random_state=seed)
model=LogisticRegression(max_iter=3000)
results=cross_val_score(model,X,y,cv=kfold)
print(results)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0,results.std()*100))
```
